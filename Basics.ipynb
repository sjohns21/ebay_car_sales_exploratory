```python
labels=autos['brand'].value_counts().index[:20]
```

```python
di={}
for label in labels:
    di[label]=autos[autos['brand']==label]['price'].mean().astype(int)
print(di)
mean_prices=di
```

the mean brand prices reflect what one would expect. the higher ends brands tend to have higher prices than lower end brands.

```python
di={}
for label in labels:
    di[label]=autos[autos['brand']==label]['odometer_km'].mean().astype(int)
print(di)
mileage=di
```

```python
prices=pd.Series(mean_prices,name='prices').sort_values()
```

```python
mileage=pd.Series(mileage,name='mileage')
```

```python
top_20=pd.DataFrame({'mean price':prices})
```

```python
top_20['mean price']
```

```python
top_20['mean mileage']=mileage
```
